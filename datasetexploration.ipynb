```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("san_pedro_cases_rounded.csv")

# Convert date column to datetime
df["date"] = pd.to_datetime(df["date"])

# Sort by date
df = df.sort_values("date")

# Plot the line graph
plt.figure(figsize=(12, 6))
sns.lineplot(x="date", y="count_Exposed", data=df, label="Exposed", linewidth=2)
sns.lineplot(x="date", y="count_Infectious", data=df, label="Infectious", linewidth=2)
sns.lineplot(x="date", y="count_Recovered", data=df, label="Recovered", linewidth=2)

# Formatting
plt.xlabel("Date")
plt.ylabel("Count")
plt.title("Trends of Exposed, Infectious, and Recovered Cases Over Time")
plt.xticks(rotation=45)
plt.legend()
plt.grid()

# Show the plot
plt.show()
```
